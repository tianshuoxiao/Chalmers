## Task 1

### Step 1

```python
import pandas as pd
```

```python
# Read train data into dataframe
train_data_path = "data/adult_train.csv"
df_train = pd.read_csv(train_data_path)
```

```python
df_train.head()
```

```python
# Read test data into dataframe
test_data_path = "data/adult_test.csv"
df_test = pd.read_csv(test_data_path)

assert all(df_test.columns == df_train.columns) # Make sure that test and train have the same columns
```

```python
# Split test and train dataframes each into X and Y dataframes
target_cols = ["target"]
feature_cols = list(set(df_train.columns).difference(target_cols)) 
```

```python
print("Targets: ", target_cols)
print("Features: ", feature_cols)
```

Split the training and testing data into Xs and Ys.

```python
df_X_train, df_Y_train = df_train[feature_cols], df_train[target_cols]
df_X_test, df_Y_test = df_test[feature_cols], df_test[target_cols]
```

### Step 2

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dicts_X_train = df_X_train.to_dict('records')
dicts_X_test = df_X_test.to_dict('records')

dv = DictVectorizer()
enc_X_train = dv.fit_transform(dicts_X_train)
```

Let's just peek at the contents :)

```python
print(enc_X_train[1])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
decision_tree_hps = {
    "max_depth": 5,
    "random_state": 42
}
decision_tree = DecisionTreeClassifier(**decision_tree_hps) # We skip tuning the hyperparameters

cv_scores = cross_val_score(decision_tree, enc_X_train, df_Y_train, scoring='accuracy')
```

```python
# Refit tree on full training set
decision_tree.fit(enc_X_train, df_Y_train)
```

```python
def agg_scores(scores):
    return sum(scores) / len(scores)
```

```python
print("Aggregated training score for default decision tree:", agg_scores(cv_scores))
```

Let's calculate the score on the test set.

```python
enc_X_test = dv.transform(dicts_X_test)
decision_tree_test_Y = decision_tree.predict(enc_X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
decision_tree_test_score = accuracy_score(decision_tree_test_Y, df_Y_test)
print("Test score for default decision tree:", decision_tree_test_score)
```

### Step 3

```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer
```

```python
decision_tree_pipeline = make_pipeline(
  FunctionTransformer(lambda df: df.to_dict('records')),
  DictVectorizer(),
  DecisionTreeClassifier(**decision_tree_hps)
)
```

If the pipeline works properly we expect to get the same test accuracy as in [step 2](#step-2).

```python
decision_tree_pipeline.fit(df_X_train, df_Y_train)
```

```python
decision_tree_pipeline_test_predictions = decision_tree_pipeline.predict(df_X_test)
decision_tree_pipeline_test_score = accuracy_score(decision_tree_pipeline_test_predictions, df_Y_test)
```

```python
print("Pipeline test score:", decision_tree_pipeline_test_score)
assert decision_tree_pipeline_test_score == decision_tree_test_score, "Not same score as the non-pipeline model!"
print("We got the same score as the non-pipeline model, success!")
```

## Task 2

### Underfitting and overfitting in decision tree classifiers

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import accuracy_score
```

```python
def todict(df):
    return df.to_dict('records')

funct = FunctionTransformer(todict)
dictv = DictVectorizer()

pipeline = Pipeline([
    ("funcT", funct),
    ("dictV", dictv),
    ("tree", DecisionTreeClassifier(random_state=42))
], memory="pipeline_cache")
```

```python
# Helper to parallelize 'map' operation
def try_parallel_map(func, params):
    try:
        from joblib import Parallel, delayed, parallel_backend
        import multiprocessing

        with parallel_backend('multiprocessing'):
            result = Parallel()(
                    delayed(func)(*param_list) for param_list in params
                )
    except Exception as e:
        # If we can't we just do it as usual
        print("Could not run in paralell. Error:", e)
        result = list(map(func, params))

    return result

def unpack_dict_list(l, key):
    return [ d[key] for d in l ]

def ignore_warnings(func):
    import warnings
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        return func()
    
```

```python
# Must start at 1 since the DecisionTreeClassifier only accepts max_depth >= 1
max_depths = list(range(1, 12))

def fit_model(max_depth):
    pipeline.set_params(tree__max_depth=max_depth)
    pipeline.fit(df_X_train, df_Y_train)

    train_pred = pipeline.predict(df_X_train)
    train_score = accuracy_score(train_pred, df_Y_train)

    test_pred = pipeline.predict(df_X_test)
    test_score = accuracy_score(test_pred, df_Y_test)

    return {
        "train_score": train_score,
        "test_score": test_score,
        "max_depth": max_depth
    }
```

```python
results = ignore_warnings(lambda:
    # Must wrap max_depths as tuple to use with try_parallel_map
    try_parallel_map(fit_model, map(lambda v: (v, ), max_depths))
)
```

```python
# Unpack results
test_scores = unpack_dict_list(results, "test_score")
train_scores = unpack_dict_list(results, "train_score")
max_depths = unpack_dict_list(results, "max_depth")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.ylabel("Accuracy")
plt.xlabel("max_depth")
plt.plot(max_depths, train_scores)
plt.plot(max_depths, test_scores)
plt.legend(["Train", "Test"])
plt.show()
```

We can see a similar effect as we did in Assignment 1. The tree is underfit up until about max_depth=6. After that, the test and train accuracy deviates.
As a max_depth=10 it looks like the test accuracy peaks and then goes down, singnifying the tree is overfit at max_depth > 10.

### Underfitting and overfitting in random forest classifiers.

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction import DictVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import accuracy_score
```

```python
pipeline = Pipeline([
    ("funcT", funct),
    ("dictV", dictv),
    ("forest", RandomForestClassifier(random_state=42, n_jobs=-1))
], memory="pipeline_cache")
```

```python
max_depths = list(range(1, 12))
ensemble_sizes = [1] + list(range(10, 50, 10)) + list(range(50, 450, 75))

def run_fit(max_depth, ensemble_size):
    pipeline.set_params(forest__max_depth=max_depth, forest__n_estimators=ensemble_size)
    pipeline.fit(df_X_train, df_Y_train)

    train_pred = pipeline.predict(df_X_train)
    train_score = accuracy_score(train_pred, df_Y_train)

    test_pred = pipeline.predict(df_X_test)
    test_score = accuracy_score(test_pred, df_Y_test)

    return {
        "max_depth": max_depth,
        "ensemble_size": ensemble_size,
        "train_score": train_score,
        "test_score": test_score
    }
```

```python
from itertools import product
import warnings

params = product(max_depths, ensemble_sizes)

results = ignore_warnings(lambda:
    try_parallel_map(run_fit, params)
)

print("First result:", results[1])
```

```python
from itertools import groupby

# Sort to make sure grouping is sorted
results.sort(key=lambda result: result['ensemble_size'])
print(results[:1])

results_dict = dict()
for key, value in groupby(results, lambda result: result['ensemble_size']):
    results_dict[key] = list(value)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(len(results_dict), 1, sharey=True, sharex=True, figsize=(15, 4*len(results_dict)))
plt.subplots_adjust(hspace=0.25)

for idx, (ensemble_size, group) in enumerate(results_dict.items()):
    # Unpack data
    max_depths = unpack_dict_list(group, "max_depth")
    train_scores = unpack_dict_list(group, "train_score")
    test_scores = unpack_dict_list(group, "test_score")

    # Plot data
    axis = axes[idx]
    axis.set_title(f"Accuracy for different max depths and n_estimators = {ensemble_size}")
    axis.set_ylabel("Accuracy")
    axis.set_xlabel("max_depth")
    axis.plot(max_depths, train_scores)
    axis.plot(max_depths, test_scores)
    axis.legend(["Train", "Test"])

    best_test_score = np.max(test_scores)
    best_test_max_depth = max_depths[np.argmax(test_scores)]
    axis.annotate(
        f'Best test score ({best_test_score:.{2}%})', 
        xy=(best_test_max_depth, best_test_score), 
        textcoords='offset points', 
        xytext=(-110, -50),
        arrowprops={"arrowstyle": "->"}
        )

plt.show()
```

```python
# Set parameters to save some time training
pipeline.set_params(forest__max_depth=10, forest__n_estimators=10)

# Refit the pipeline to be able to get the calculated feature importance
ignore_warnings(lambda:
    pipeline.fit(df_X_train, df_Y_train)
)
```

```python
feature_importances = pipeline.named_steps["forest"].feature_importances_
feature_names = pipeline.named_steps["dictV"].feature_names_
```

```python
feature_importances_per_feature = list(zip(feature_names, feature_importances))
# Sort by feature importance
feature_importances_per_feature.sort(reverse=True, key=lambda pair: pair[1])
```

```python
# Print top feature scores
for name, importance in feature_importances_per_feature[:5]:
    print(f"Feature {name} has importance {importance}")
```
